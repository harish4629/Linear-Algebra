#Q56
Explain the concept of Exception Handling in Python with suitable program.

Answer:

An exception is a Python object that represents an error.
If you have some suspicious code that may raise an exception, you can defend your program by placing the suspicious code in a try: block. After the try: block, include an except: statement, followed by a block of code which handles the problem as elegantly as possible.
#Example
try:
   fh = open("testfile", "w")
   fh.write("This is my test file for exception handling!!")
except IOError:
   print ("Error: can\'t find file or read data")
else:
   print ("Written content in the file successfully")
   fh.close()

2)Read a file line by line using readline()
lineStr = fileHandler.readline()
# Open file        
fileHandler = open ("data.txt", "r")
while True:
    # Get next line from file
    line = fileHandler.readline()
    # If line is empty then end of file reached
    if not line :

3)Read file line by line with context manager (with block)
# Open file 
with open ("data.txt", "r") as fileHandler:
    # Read each line in loop
    for line in fileHandler:
        # As each line (except last one) will contain new line character, so strip that
        print(line.strip())
4)Get List of lines in file with context manager (with block)

with open ("data.txt", "r") as myfile:

ii) Different betn append and write mode
Ans- we have a two mode in which we open the file 
'w' open file in write mode.
'a' open file in append mode.
- Both mode write the content in the file, if file not exist then they create that file.
- for a existing file, 
- write mode 'w' erase the previous data of the file and insert new data while append mode don't erase previous data, it append new data with previous one.
